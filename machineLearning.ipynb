```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Create a list of lists containing the data for each row
data = [
    [1, 2, 3, 10000],
    [2000, 3000, 4000, 20000],
    [5000, 6000, 7000, 30000],
]

# Create a DataFrame from the list of lists
dataset = pd.DataFrame(data, columns=['Facebook Ads', 'Google Ads', 'Instagram Ads', 'Revenus'])

# Extract the input and output variables
X = dataset[['Facebook Ads', 'Google Ads', 'Instagram Ads']]
y = dataset['Revenus']

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions and evaluate the model
y_predt = model.predict(X)

# ... (rest of the code

plt.figure(figsize=(10, 6))
plt.scatter(y, y_predt)
plt.plot([min(y), max(y)], [min(y), max(y)], color='red')
plt.xlabel('Revenu observé')
plt.ylabel('Revenu prédits')
plt.title('Régression linéaire multiples')
plt.show()

coefficients = model.coef_
intercept_all = model.intercept_
equation_all = f"Revenu = {intercept_all:.2f}"
for i, col in enumerate(X.columns):
    equation_all += f"+({coefficients[i]:.2f} * {col})"
print(f"équation de la regression linéaire multiple: \n {equation_all}")
```

```python
# Least square
import statsmodels.formula.api as smf

model = smf.ols(formula='Revenus ~ Facebooks', data=df1).fit()

# Here are estimated intercept and slope by least square estimation
# Attribute 'params' returns a list of estimated parameters form model
b0_ols = model.params[0]
b1_ols = model.params[1]

df1['BestResponse'] = b0_ols + b1_ols*df1['Facebooks']

# Also want to know the error of of guess...
df1['error'] = df1['Revenus'] - df1['BestResponse']


# plot your estimated line together with the points
plt.figure(figsize=(10, 10))
# See if the error drops after you use least square method
plt.title('Sum of sqaured error is {}'.format((((df1['error'])**2)).sum()))
plt.scatter(df1['Facebooks'], df1['Revenus'], color='g', label='Observed')
plt.plot(df1['Facebooks'], df1['GuessResponse'], color='red', label='GuessResponse')
plt.plot(df1['Facebooks'], df1['BestResponse'], color='blue', label='BestResponse')
plt.legend()
plt.xlim(df1['Facebooks'].min()-2, df1['Facebooks'].max()+2)
plt.ylim(df1['Revenus'].min()-2, df1['Revenus'].max()+2)
plt.show()
```
